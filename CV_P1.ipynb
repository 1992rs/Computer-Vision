```python
#Let's extract the three channels
Red_channel, Green_channel, Blue_channel = rose_image_cv[:,:,0], rose_image_cv[:,:,1],rose_image_cv[:,:,2]
```

```python
# Formula to convert into Grey Scale
rose_grey_scale = (0.299 * Red_channel + 0.587 * Green_channel + 0.114 * Blue_channel)
```

```python
rose_greyscale_image_resized = cv2.resize(rose_grey_scale,new_dim,interpolation = cv2.INTER_AREA) 
plt.imshow(rose_greyscale_image_resized,cmap='gray');
```

```python
# Converting to HSI

rose_HSV = cv2.cvtColor(rose_image_cv,cv2.COLOR_RGB2HSV)

# displaying the image in HSV format
plt.imshow(rose_HSV);

# Save the HSV image
cv2.imwrite('HSV_Image.png', rose_HSV)
```

```python
# Converting the image to YCrCb

rose_YCrCb = cv2.cvtColor(rose_image_cv,cv2.COLOR_RGB2YCrCb)

plt.imshow(rose_YCrCb);

# Save the YCrCb image
cv2.imwrite('YCrCb_Image.png', rose_YCrCb)
```

```python
# Let's divide the original image into 2 parts vertically

h,w,c = rose_image_cv.shape

print(f'Original image || Image height: {h}')
print(f'Original image || Image width: {w}')
print(f'Original image || Number of channels: {c}')
```

```python
# Notice that if we are cutting an image vertically then we have to divide the width
cut_point = round(w/2)
print(f'Cut point : {cut_point}')
```

```python
left_side_image = rose_image_cv[:, :cut_point]
right_side_image = rose_image_cv[:, cut_point:]
```

```python
plt.imshow(left_side_image);
```

```python
plt.imshow(right_side_image);
```

```python
# Plot the images
fig=plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(left_side_image)
ax1.set_title('left_side_image Image')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(right_side_image)
ax2.set_title('right_side_image Image')
plt.show();
```

```python
# Cut the left side image into two parts horizonatlly

# Let's divide the original image into 2 parts vertically

hl,wl,cl = left_side_image.shape

cut_point = round(hl/2)
print(f'Cut point : {cut_point}')

left_upper_side_image = left_side_image[:cut_point,]
left_bottom_side_image = left_side_image[cut_point:,:]
```

```python
plt.imshow(left_upper_side_image);
```

```python
plt.imshow(left_bottom_side_image);
```

```python
# Cut the left side image into two parts horizonatlly

# Let's divide the original image into 2 parts vertically

hl,wl,cl = right_side_image.shape

cut_point = round(hl/2)
print(f'Cut point : {cut_point}')

right_upper_side_image = right_side_image[:cut_point,]
right_bottom_side_image = right_side_image[cut_point:,:]
```

```python
plt.imshow(right_upper_side_image);
```

```python
plt.imshow(right_bottom_side_image);
```

```python
# Attach these 4 parts to make the complete image again

# Attach the left and right side upper images 
semi_upper_rose = cv2.hconcat([left_upper_side_image, right_upper_side_image])
  
# check the image
plt.imshow(semi_upper_rose);
```

```python


# Attach the left and right side bottom images 
semi_lower_rose = cv2.hconcat([left_bottom_side_image, right_bottom_side_image])
  
# check the image
plt.imshow(semi_lower_rose);
```

```python
# Attch these two halves to get the full image

reconstructed_rose = cv2.vconcat([semi_upper_rose, semi_lower_rose])

# check the image
plt.imshow(reconstructed_rose);
```

```python
print(f'Original image || Shape of this array is:{rose_image_cv.shape}')
print(f'Reconstructed image || Shape of this array is:{reconstructed_rose.shape}')
```

```python
rose_red_removed = rose_image_cv.copy() # Make a copy of original image

#Isolate the areas where the color is black(every channel=0) and white (every channel=255)
red_pixels=np.where((rose_red_removed[:,:,0]==255) & (rose_red_removed[:,:,1]==0) & (rose_red_removed[:,:,2]==0))

#Turn red pixels to white
rose_red_removed[red_pixels]=(255,255,255)

# Plot the images
fig=plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(rose_image_cv)
ax1.set_title('Original Image')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(rose_red_removed)
ax2.set_title('Reconstructed Image')
plt.show();
```

```python
# Traverse through all the pixels of the image and remove all of the red intensities.

rose_red_removed = rose_image_cv.copy() # Make a copy of original image

# Now remove all red with 0
rose_red_removed[np.where(rose_red_removed==255)] = 0

plt.imshow(rose_red_removed);
```

```python
# Load original image
rose_image_cv = cv2.imread(image_path)

# We define colors for openCV (BGR format)
red   = [0,0,255]
green = [0,255,0]
blue  = [255,0,0]
white = [255,255,255]
black = [0,0,0]

# Remove all red from image
rose_red_removed = rose_image_cv.copy()

# Find all the red pixels (255,0,0) and convert them with (0,0,0)
rose_red_removed[np.all(rose_red_removed == red, axis=-1)] = black

plt.imshow(rose_red_removed);
```

```python
# For the original image, change the blue intensities with red, red with green, and green with blue.

# B-R ; R-G ; G-B
# BRG to RGB

rose_chnaged_pixels = rose_image_cv.copy() # Make a copy of original image

# Blue goes to red
rose_chnaged_pixels[np.all(rose_chnaged_pixels == blue, axis=-1)] = red

plt.imshow(rose_chnaged_pixels);
```

```python
rose_chnaged_pixels = rose_image_cv.copy() # Make a copy of original image

# red goes to green
rose_chnaged_pixels[np.all(rose_chnaged_pixels == red, axis=-1)] = green

plt.imshow(rose_chnaged_pixels);
```

```python
rose_chnaged_pixels = rose_image_cv.copy() # Make a copy of original image

# green goes to blue
rose_chnaged_pixels[np.all(rose_chnaged_pixels == green, axis=-1)] = blue

plt.imshow(rose_chnaged_pixels);
```
