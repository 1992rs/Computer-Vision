```python
op_xor = cv2.bitwise_xor(img_1, img_2, mask = None)
plt.imshow(op_xor);
```

```python
op_not1 = cv2.bitwise_not(img_1, mask = None)
op_not2 = cv2.bitwise_not(img_2, mask = None)

# Plot the images
fig=plt.figure(figsize = (10,8))

ax1 = fig.add_subplot(1,2,1)
ax1.imshow(op_not1)
ax1.set_title('Image-1')

ax2 = fig.add_subplot(1,2,2)
ax2.imshow(op_not2)
ax2.set_title('Image-2')

plt.show();
```

###  Scaling, Rotating, Shifting 

```python
img_1 = cv2.imread(folder_path + '/rose.jpeg')
```

```python
# Height and width of image
(height, width) = img_1.shape[:2]
```

```python
# Scale image to a smaller image
scaled_image = cv2.resize(img_1, (int(width / 20), int(height / 20)), interpolation = cv2.INTER_CUBIC)
```

```python
img_1_rgb = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
scaled_image_rgb = cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB)
```

```python
# Plot the images
fig=plt.figure(figsize = (10,8))

ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img_1_rgb)
ax1.set_title('Image')

ax2 = fig.add_subplot(1,2,2)
ax2.imshow(scaled_image_rgb)
ax2.set_title('Scaled Image')

plt.show();
```

```python
print(f'Original image || Shape of this array is:{img_1.shape}')
print(f'Scaled image || Shape of this array is:{scaled_image.shape}')
```

```python
2048/20,1298/20
```

```python
rotation_matrix = cv2.getRotationMatrix2D((height / 2, width / 2), 180, 1)
rotated_image = cv2.warpAffine(img_1, rotation_matrix, (height, width))

M = np.float32([[1, 0, 100], [0, 1, 50]])
translated_image = cv2.warpAffine(img_1, M, (height, width))
```

```python
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)
translated_image_rgb = cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)
```

```python
# Plot the images
fig=plt.figure(figsize = (10,8))

ax1 = fig.add_subplot(1,3,1)
ax1.imshow(img_1_rgb)
ax1.set_title('Image')

ax2 = fig.add_subplot(1,3,2)
ax2.imshow(rotated_image_rgb)
ax2.set_title('Rotated Image')

ax2 = fig.add_subplot(1,3,3)
ax2.imshow(translated_image_rgb)
ax2.set_title('Translated Image')

plt.show();
```

```python
print(f'Original image || Shape of this array is:{img_1.shape}')
print(f'Scaled image || Shape of this array is:{rotated_image.shape}')
print(f'Scaled image || Shape of this array is:{translated_image.shape}')
```
